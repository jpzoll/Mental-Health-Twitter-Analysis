# K-Means Clustering
## Spiritual Accounts & Actor Accounts

```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
from sklearn.cluster import KMeans
import numpy as np
```

```python
df = pd.read_csv('data/moods_and_community_moods.csv')
```

```python
df['group'] = df.index.to_frame()
df['group'] = df['group'].apply(lambda group: 1 if group <= 38 else 2)
```

```python
df['mood'] = df['mood'].fillna(np.mean(df.mood.iloc[22:]))
df['community_mood'] = df['community_mood'].fillna(np.mean(df.community_mood.iloc[22:]))
```

```python
df.head()
```

```python
plt.scatter(df['mood'], df['community_mood'])
```

```python
kmeans = KMeans(n_clusters=2)
kmeans
```

```python
cluster_values = list(kmeans.fit_predict(df[['mood', 'community_mood']]))
df['cluster'] = cluster_values
```

```python
df.head(3)
```
